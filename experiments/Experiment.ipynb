```python
scenesPromptTemplate = "Below is a movie script, please generate various scenes for this script. Each scene should start with heading ##Scene.\n\n {script}"
scenesChain = (
    ChatPromptTemplate.from_template(scenesPromptTemplate)
    | model
    | StrOutputParser()
)

scenesRunOutput = scenesChain.invoke(runOutput["Script"])
print(scenesRunOutput)

```

```python
print(scenesRunOutput)
```

```python
scenesList = scenesRunOutput.split('##Scene')[1:]

```

```python
import re
scenesList = [re.sub(r"^\s*\d+:\s*", "", s) for s in scenesList]

scenesList
```

```python
# Scene to Frame prompt

framesPromptWithScript = "Below is a Movie Script followed by one Scene Description from the script. Looking at the Scene Description, think of 3-4 frames and create image generation prompts for the scene. \n\n Start of Movie Script\n\n {script} \n\n End of Movie Script\n\n Start of Scene Description\n\n{scene}\n\n End of Scene Description."
framesPromptWOScript = "Below is a Scene Description. Looking at the Scene Description, think of 3-4 frames and create image generation prompts for these frames, only provide the image prompts without frame description. \n\n {scene}\n\n "
sceneFramesChain = (
    ChatPromptTemplate.from_template(framesPromptWOScript)
    | model
    | StrOutputParser()
)

framesPromptResult = sceneFramesChain.batch([{"scene":value} for value in scenesList])
```
